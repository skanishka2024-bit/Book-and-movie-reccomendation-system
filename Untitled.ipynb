```python
movies_df = pd.read_csv("ml-latest-small/movies.csv")
books_df = pd.read_csv("Book datasets/Goodreadss Books.csv")
```

```python
books_df.drop(columns=['Unnamed: 0',   'author', 'series', 'description',
        'awards', 'characters', 'places', 'isbn', 'isbn13',
       'language', 'first_publish_date', 'publish_date', 'num_pages',
       'num_ratings', 'num_reviews', 'rated_1', 'rated_2',
       'rated_3', 'rated_4', 'rated_5'],inplace=True)
```

```python
MENU = ("Art,Biography,Business,Children's,Christian,Classics,Comics,Cookbooks,"
        "Ebooks,Fantasy,Fiction,Graphic Novels,Historical Fiction,History,Horror,"
        "Memoir,Music,Mystery,Nonfiction,Poetry,Psychology,Romance,Science,"
        "Science Fiction,Self Help,Sports,Thriller,Travel,Young Adult")

def strip_menu_boilerplate(raw):
    text = str(raw)
    doubled = MENU + "," + MENU
    if text.startswith(doubled):
        text = text[len(doubled):]
    elif text.startswith(MENU):
        text = text[len(MENU):]
    return text.lstrip(',').strip()

books_df['genres'] = books_df['genres'].apply(strip_menu_boilerplate)
```

```python
movies_df['genres'] = movies_df['genres'].fillna('')
movies_df['genres'].sample()
```

```python
movies_df.columns
```

```python
books_df['genres'] = books_df['genres'].fillna('')
```

```python
movies_df['genre_list'] = movies_df['genres'].apply(
    lambda x: [g.strip().lower() for g in str(x).split('|') if g.strip()] if pd.notna(x) else []
)
books_df['genre_list'] = books_df['genres'].apply(
    lambda x: [g.strip().lower() for g in str(x).split(',') if g.strip()] if pd.notna(x) else []
)

movies_df = movies_df[movies_df['genre_list'].map(len) > 0].reset_index(drop=True)
books_df = books_df[books_df['genre_list'].map(len) > 0].reset_index(drop=True)
```

```python

from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
mlb.fit(pd.concat([movies_df['genre_list'], books_df['genre_list']]))

movie_vectors = mlb.transform(movies_df['genre_list'])
book_vectors = mlb.transform(books_df['genre_list'])

print(movie_vectors.shape, book_vectors.shape)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

te = TransactionEncoder()
te_ary = te.fit(movies_df['genre_list']).transform(movies_df['genre_list'])
onehot_df = pd.DataFrame(te_ary, columns=te.columns_)

frequent_itemsets = apriori(onehot_df, min_support=0.03, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

```

```python
import pickle

with open('movies_df.pkl', 'wb') as f:
    pickle.dump(movies_df, f)

with open('books_df.pkl', 'wb') as f:
    pickle.dump(books_df, f)
with open('movie_vectors.pkl', 'wb') as f:
    pickle.dump(movie_vectors, f)
with open('book_vectors.pkl', 'wb') as f:
    pickle.dump(book_vectors, f)

with open('genre_association_rules.pkl', 'wb') as f:
    pickle.dump(rules, f)

```
